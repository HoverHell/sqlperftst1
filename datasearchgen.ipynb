```python
import random
import collections
import itertools
```

```python
all_words = open('/usr/share/dict/words').read().decode('utf-8').splitlines()
all_words = (w.strip() for w in all_words)
all_words = (w for w in all_words if not w.endswith("'s"))
all_words = (w.lower() for w in all_words)
all_words = list(all_words)
all_words = random.sample(all_words, int(1e4))
## For simplicity, make the likelihood alphabetical
all_words.sort()
```

```python
pkgens = collections.defaultdict(lambda: itertools.count(1))

def mk_node():
    length = random.randrange(2, 30)
    # words = random.sample(all_words, length)
    ## Not an equiprobable distribution, not necessarily unique either:

    def select_word(src):
        mean = len(src) / 1e3
        idx = int(random.expovariate(1.0 / mean))
        idx = min(idx, len(src) - 1)
        return src[idx]

    words = [select_word(all_words) for _ in xrange(length)]
    node = dict(
        id=next(pkgens['node']),
        loc=' '.join(words),
        weight=random.expovariate(0.1),
    )
    return node

def node2flts(node):
    words = node['loc'].split()
    flts = [
        dict(
            id=next(pkgens['filt']),
            node_id=node['id'],
            word=word,
        )
        for word in set(words)
    ]
    return flts
```

```python
nodes = [mk_node() for _ in xrange(int(2e5))]
flts = (flt for node in nodes for flt in node2flts(node))
```

```python
schema_sql = '''
DROP TABLE IF EXISTS "filt";
DROP TABLE IF EXISTS "node";

CREATE TABLE "node" (
    "id" integer PRIMARY KEY,
    "loc" varchar(255) NOT NULL,
    "weight" real
);
CREATE INDEX "node_loc" ON "node" ("loc");
CREATE INDEX "node_weight" on "node" ("weight");

CREATE TABLE "filt" (
    "id" integer PRIMARY KEY,
    "node_id" integer NOT NULL,
    "word" varchar(120) NOT NULL
);
CREATE INDEX "filt_word" ON "filt" ("word");
CREATE INDEX "filt_node_id" ON "filt" ("node_id");
'''

sql = [schema_sql]
sql.append('''BEGIN TRANSACTION;''')

## Generating an SQL file here, from trusted input, so no point in using params
node_sql = '''insert into node (`id`, `loc`, `weight`) values (%s);'''
flt_sql = '''insert into filt (`id`, `node_id`, `word`) values (%s);'''

def _escape_val(val):
    if isinstance(val, basestring):
        if "'" in val:
            return u"X'%s'" % (val.encode('utf-8').encode('hex'),)
        return "'%s'" % (val,)
        return val.replace("'", "''")
    return val


def _escape_dct(dct):
    return {k: _escape_val(v) for k, v in dct.items()}


for node in nodes:
    params = "%(id)d, %(loc)s, %(weight).1f" % _escape_dct(node)
    sql.append(node_sql % (params,))

for flt in flts:
    params = "%(id)d, %(node_id)d, %(word)s" % _escape_dct(flt)
    sql.append(flt_sql % (params,))
    
sql.append('COMMIT;')

sql_s = u'\n'.join(sql) + '\n'
with open('/home/hhell/perf1t.sql', 'w') as f:
    f.write(sql_s.encode('utf-8'))
```
